```python
import numpy as np
import matplotlib.pyplot as plt
import time
```

### Hledač kořenů

```python
def bisection(f, a0, b0, eps=1E-10):
    a, b = a0, b0
    while b - a > 2*eps:
        x = (a + b)/2
        if f(a)*f(x) < 0:
            b = x
        else:
            a = x
    return x

def newton(f, a0, b0, delta=1E-10):
    a, b = a0, b0
    def df(x, h = 1E-10):
        return (f(x+h)-f(x-h))/(2*h)
    xnew = (a+b)/2
    xold = a
    while abs(xnew - xold) > delta:
        xold = xnew
        xnew = xold - f(xold)/df(xold)
    return xnew

def plot_result(f, a0, b0, x, title):
    print("root: ", x)
    plt.title(title)
    plt.plot(x, f(x), "bo")                                               # root
    plt.plot(np.linspace(a0, b0, 100), f(np.linspace(a0, b0, 100)), "r")  # function
    plt.axhline(y=0)
    plt.show()
```

### Data

#### Funkce 1

```python
funkce1 = lambda x: x**4 + 5*x**3 + x**2 -9
funkce2 = lambda x: 2*x - np.exp(-3*x) + 8
funkce3 = lambda x: np.sin(x-9)
```

```python
a1, b1 = -1.2, 3.2
plot_result(funkce1, a1, b1, bisection(funkce1, a1, b1), "Bisection: x^4 + 5x^3 + x^2 - 9")
plot_result(funkce1, a1, b1, newton(funkce1, a1, b1), "Newton: x^4 + 5x^3 + x^2 - 9")

old_time = time.time()
lambda: bisection(funkce1, a1, b1)
print("bisection trval: " + str(time.time() - old_time) + "s")
old_time = time.time()
lambda: newton(funkce1, a1, b1)
print("newton trval: " + str(time.time() - old_time) + "s")
```

#### Funkce 2

```python
a2, b2 = -1, 1
plot_result(funkce2, a2, b2, bisection(funkce2, a2, b2), "Bisection: 2x - e^(-3x) + 8")
plot_result(funkce2, a2, b2, newton(funkce2, a2, b2), "Newton: 2x - e^(-3x) + 8")

old_time = time.time()
lambda: bisection(funkce2, a2, b2)
print("bisection trval: " + str(time.time() - old_time) + "s")
old_time = time.time()
lambda: newton(funkce2, a2, b2)
print("newton trval: " + str(time.time() - old_time) + "s")
```
